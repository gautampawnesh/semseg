```python
%matplotlib inline
import json 
import numpy as np
import matplotlib.pyplot as plt
```

```python
universal_f = "class_mapping/universal_classes.csv"
```

```python
import pandas as pd
```

```python
u_data = pd.read_csv(universal_f, delimiter=";")
```

```python
cls = u_data["class_id"]
```

```python
new_cls = u_data["Unnamed: 3"]
```

```python
len(new_cls)
```

```python
u_cls_new_cls = dict(zip(cls, new_cls))
```

```python
input_file = "class_mapping/playing_for_data_class_mapping.csv"
```

```python
dataset_f = input_file
```

```python
d_data = pd.read_csv(dataset_f, delimiter=";")
```

```python
d_cls = d_data["universal_class_id"]
```

```python
d_cls_new_cls = [u_cls_new_cls[k] for k in d_cls] 
```

```python
d_data["universal_class_id"] = d_cls_new_cls
```

```python
d_data
```

```python
# universal to data
```

```python
d_data["universal_to_dataset"].tolist()
```

```python
ll = []
for k in d_data["universal_to_dataset"].tolist():
    if pd.isna(k):
        ll.append(np.nan)
        continue
    k_i = k.split(",")
    temp_str = ""
    for i in k_i:
        i = int(i)
        new_i = u_cls_new_cls[i]
        if len(k_i) >1 and new_i ==0:
            continue
        if temp_str == "":
            temp_str += str(new_i)
        else:
            temp_str += "," + str(new_i)
    ll.append(temp_str)
```

```python
ll
```

```python
d_data["universal_to_dataset"] = ll
```

```python
header = d_data.columns.tolist()
```

```python
d_data.to_csv(input_file, sep=";", index=False)
```
